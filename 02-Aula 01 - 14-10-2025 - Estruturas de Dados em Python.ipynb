### $ \LaTeX\ $

É possível escrever fórmulas do latex em seus notebooks.

Para escrever expressões matemáticas ao longo do texto, basta usar \$ entre os comando do Latex. Exemplo: $ a = \delta + \alpha + \beta $.

Para escrever expressões matemáticas centralizadas, basta escrevê-las entre \$\$, exemplo:

$$ f(x) = \frac{y}{2}.$$
