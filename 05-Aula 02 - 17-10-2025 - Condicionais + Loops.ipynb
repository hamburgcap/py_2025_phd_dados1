## Range

 - O resultado gerado é uma lista que vai de 0 até o número dado como argumento

 - Só recebendo 1 argumento: o final. Nesse caso o range entende que o início é o zero e o tamanho do passo é 1.

 - Range representa uma sequência imutável de números e é usado para fazer um número específico de loops nas estruturas de repetição.

- Recebe de 1 a 3 argumentos: início, final (aberto) e tamanho do passo

- o Range sozinho não produz a lista de números. Para tanto, é necessário transformá-lo em um iterável (exceto dicionário)

```
a = range(10)
print(a)
```

```
a = list(range(10))
print(a)
```

```
a = tuple(range(10))
print(a)
```

```
a = set(range(10))
print(a)
```

## Condicionais

Um bloco condicional pode conter:
- 1 **'if'**

- 0 ou 1 **'else'**

- 0,1 ou múltiplos **'elif'**

O comando **'if'**

```
ipca_2020 = 0.025
print(ipca_2020)
```

```
if ipca_2020 < 0.04:
  print('A inflação está abaixo da meta.')
  print('A inflação realmente está abaixo da meta.')
```

```
ipca_2020 = 0.05
```

Usando o 'else'

```
print(ipca_2020)
if ipca_2020 > 0.04:
  print('A inflação está acima da meta.')
else:
  print('A inflação está na meta ou abaixo dela.')
```

```
ipca_2020 = 0.05
```

Tornando mais elegante

```
ipca_2020 = 0.05
if ipca_2020 > 0.04:
  print(f'A inflação de {ipca_2020*100}% está acima da meta estabelecida pelo CMN de 4% para 2020.')
elif ipca_2020 == 0.04:
  print(f'A inflação de {ipca_2020*100}% está na meta estabelecida pelo CMN para 2020.')
else:
  print(f'A inflação de {ipca_2020*100}% está abaixo da meta estabelecida pelo CMN de 4% para 2020.')
```

### Proposta do Sandro

```
#problema do Marlon
valoremp = 100000
taxa = 0.68
prazo = 60

saldodevedoranterior[0] = valoremp
for i in range(0,prazo):
  jurospagos[i] = saldodevedoranterior[i]*taxa/100
  amortizacao[i] = 100
  parcelas[i] = amortizacao[i] + jurospagos[i]
  saldodevedorfinal[i] = saldodevedoranterior[i] - amortizacao[i]
  if i < prazo-1:
    saldodevedoranterior[i+1] = saldodevedorfinal[i]
```
