## Objetivo do curso

Preparar o aluno do Mestrado Profissional em Economia e Finanças para usar ferramentas e liderar equipes de *Data Science*  para tomada de decisões de negócios.

## Ementa do curso

Uso do Google Colab, Jupyter notebook, estrutura de dados em Python, condicionais, loops, Funções, Numpy e Aprendizado de Máquinas.
