```python
def tamanho(lista_string):

  for x in range(len(lista_string)-1):
    for j in range(len(lista_string)-x-1) :
     if len(lista_string[j])>len(lista_string[j+1]):
        temp = lista_string[j]
        lista_string[j]=lista_string[j+1]
        lista_string[j+1]=temp
  print(lista_string)

lista=['Paralelograma','Python','Testestre','Google','Corinthians','A','ABC']
tamanho(lista)
```
