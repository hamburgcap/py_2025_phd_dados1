### _break_, _continue_ and _pass_

+ O _break_ interrompe o fluxo a iteração;
+ O _continue_ volta para o fluxo normalmente, ou seja, pula a iteração;
+ O _pass_ não faz nada.

```
numbers = range(1,1000)
num_sum = 0
count = 0

for x in numbers:
    num_sum = num_sum + x
    count = count + 1
    if count == 120:
        break
print("A soma dos primeiros",count,"inteiros:", num_sum)
```

```
import time
t0 = time.time()
t0
```

```
t0 = time.time()
while 'Fluminense':
  print("Fluminense")
  t1 = time.time() - t0
  if t1 > 0.0001:
    break
print("A execução levou", t1)
```

```
bool('Fluminense')
```

```
x = 0
while True: #loop eterno
    print(x)
    x += 1
    # isto é equivalente a
    #x = x+1
    #e vale para as demais operações (-, *, /)
    if x > 10:
        break
```

A diferença entre o **pass** e o **continue**

Exemplo 1

```
for number in range(10):
    if number in [5,7,9]:
        continue
    print(number)
```

```
for number in range(10):
    if number in [5,7,9]:
        pass
    print(number)
```

Exemplo 2

```
for a in ['BBAS3','BBDC3','SANB11','BPAN4']:
    if a.endswith('3'):
        continue
    print(a)
```

```
for a in ['BBAS3','BBDC3','SANB11','BPAN4']:
    if a.endswith('3'):
        pass
    print(a)
```

### Como lidar com erros e exceções: _try_, _except_, _else_, _finally_

[Exception types in Python](https://stackabuse.com/python-exception-handling/):

+ **_try_**           tenta executar a linha de código;  
+ **_except_**        caso dê erro no try, o except é chamado;  
+ **_else_**          caso o try tenha funcionado, o else é chamado;  
+ **_finally_**       o try falhando ou não, o finally é chamado.

Como lidar com erros?

```
x = 2
y = [1,3]
x + y # error
```

```
try:
    x + y
except:
    print("Ocorreu um erro")
```

```
y = 10
```

```
try:
    x + y
except:
    print("Ocorreu um erro")
else:
    print("Está tudo ok")
```

```
try:
    x + y
except:
    print("Ocorreu um erro")
else:
    print("Está tudo ok")
finally:
    print("Fim do bloco")
```

```
def somaskol(x,y):
   try:
       z = x + y
   except:
         print('O códido ficou quadrado.')
   else:
       print('O código rodou redondo!')
       return z
   finally:
         print('É isto, pessoal!')
```

```
somaskol([1],2)
```

### Função Lambda

- Lambda é uma função anônima que aceita mais de um argumento;

- Só suporta uma expressão;

- É o ideal quando não precisamos de funções muito complexas.

Exemplo 1: Soma

```
soma1 = lambda x,y : x + y
soma1(1,2)
```

```
def soma2(x,y):
    return x+y
```

```
print(soma1(1,2))
print(soma2(1,2))
teste1 = soma2(1,2)
print(type(teste1))
print(teste1)
```

Exemplo 2: Elevar ao quadrado

```
quadrado = lambda x: x**2
```

```
teste = quadrado(2)
print(teste)
```

Testar se um número é par

```
par = lambda x:True if x%2==0 else False
```

```
par(17)
```

```
par(16)
```

### Comando zip

Esse comando recebe duas ou mais sequências não vazias e retorna uma **tupla** com os elementos de mesmo índice

```
variacao = [4.2,1.8,3.3,1.1,5.1,2.4,11.1]
indicadores = ['Produção Industrial','Vendas no Comércio','Serviços (PMS)','Fluxo pedagiado de veículos',
               'Confiança do Consumidor','Confiança Empresarial']
z = zip(variacao,indicadores)
```

```
print(z)
type(z)
```

```
list(z)
```

z é um iterador, logo precisa ser carregado de novo caso seja necessário criar de novo o objeto

```
z = zip(indicadores,variacao)
```
