### Tipo Dicionário:
Dicionários são sequências de itens na forma de pares *key:values* que se encontram dentro de chaves e separados por vírgula.

Comandos interessantes: get, items, keys, update, values

**Observações:**

**1) Keys são únicas, values podem ser repetidos**

**2) Tipos mutáveis não podem ser chaves em dicionários. Valores podem ser mutáveis**


```
dicionario = {'chave1':[1,2,3],'chave2':2,'outra chave':4, 45: 26, 'outro':23}
dicionario
```

A partir de um código que exporta um relatório do FMI, criou-se um dicionário com as previsões de PIB para vários países.

```
import urllib.request
import pandas as pd

## fazendo o download

urllib.request.urlretrieve('https://www.imf.org/~/media/Files/Publications/WEO/2020/Update/June/English/data/WEOJun2020update.ashx?la=en', "WEOJun2020update.xlsx")
```

```
df = pd.read_excel('WEOJun2020update.xlsx', sheet_name='Real GDP Growth', header = 6)
## Forma alternativa usando a posição da tabela
#df = pd.read_excel('WEOJun2020update.xlsx', sheet_name=2, header = 6)
print(df.head())
print(df.tail())
```

```
## lendo o arquivo

df = pd.read_excel('WEOJun2020update.xlsx', sheet_name=2, header = 6)

## Renomeando uma coluna

df.rename(columns = {'Unnamed: 0':'Country'}, inplace = True)

## Deletando 3 linhas (a primeira, a última e a penúltima) - Inspecionei usando df.head() e df.tail()

df.drop(df.index[[0,-1,-2]], inplace = True)

## Deletando algumas colunas sem nome

cols_unnamed = [cols_del for cols_del in df.columns if 'Unnamed' in str(cols_del)]
cols_unnamed = cols_unnamed + ['2020.1','2021.1']
print(cols_unnamed)
df.drop(cols_unnamed, axis = 1, inplace = True)

## Definindo a coluna 'Country' como index

df.set_index('Country', inplace = True)
```
