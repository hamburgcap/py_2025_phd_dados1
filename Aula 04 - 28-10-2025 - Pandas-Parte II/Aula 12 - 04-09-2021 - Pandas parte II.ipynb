Etapa 1 - Obtendo os dados

```
index_pct_change_array = []

stock_index =  ['^BVSP', '^GDAXI', '^FTSE', '^DJI', '^GSPC']

for ind in stock_index:

    index_pct_change = yf.Ticker(ind).history(start = "2020-09-07", stop = "2020-09-08")['Close'].pct_change()[1]

    index_pct_change_array = np.append(index_pct_change, index_pct_change_array)
```

**Criando um novo dataframe através de um filtro**

Nunca esquecer de usar o copy!

df_rj é um dataframe só com cidades do Rio de Janeiro

```
df_rj = df2[df2['Nome da Unidade da Federação'] == 'Rio de Janeiro'].copy()
```

### Análise: Quais cidades do Rio de Janeiro mais cresceram entre 2010 e 2017?

```
df_rj = df_rj[df_rj.Ano.isin([2010,2017])]
```

Selecionando algumas colunas

```
df_rj = df_rj[['Ano','Nome do Município','Atividade com maior valor adicionado bruto', 'Produto Interno Bruto per capita, \na preços correntes\n(R$ 1,00)']]
```

```
df_rj.rename(columns = {'Produto Interno Bruto per capita, \na preços correntes\n(R$ 1,00)':'PIB_per_capita', 'Atividade com maior valor adicionado bruto':'Principal Atividade'}, inplace = True)
# Equivalente a df_rj = df_rj.rename....
```

```
df_rj.sort_values(['Nome do Município', 'Ano'], inplace = True)
```

```
df_rj.set_index(['Nome do Município', 'Principal Atividade'], inplace = True)
```

```
df_rj['Principal Atividade'].unique()
```

```
df_rj.tail()
```

```
df_rj.groupby('Nome do Município')['PIB_per_capita'].pct_change()
```

```
pd.DataFrame(df_rj.groupby('Nome do Município')['PIB_per_capita'].pct_change())
```

```
df_rj_final = pd.DataFrame(df_rj.groupby('Nome do Município')['PIB_per_capita'].pct_change()).reset_index()
```

```
df_rj_final.head()
```

```
df_rj_final.dropna().head()
```

```
df_rj_final.groupby('Principal Atividade')['PIB_per_capita'].mean().sort_values(ascending = False)
```

```
df_rj_final.dropna().sort_values('PIB_per_capita', ascending = False).head(20)
```

```
df_rj_final.dropna().sort_values('PIB_per_capita', ascending = False).head(20).to_csv('20 Cidades que mais cresceram no Rio de Janeiro', sep = ';', encoding = 'latin-1')
```
